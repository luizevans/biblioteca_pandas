```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

### Tratamento dos Nulos

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS', inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preo'].mean()
```

```python
vendas['Preo'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas[['Preo', 'Preo_com_frete']].describe()
```

```python
preco_errado = vendas.query('Preo > Preo_com_frete')
vendas_correta = vendas.query('Preo < Preo_com_frete')
```

```python
vendas_correta[['Preo', 'Preo_com_frete']].describe()
```

```python
vendas_correta.head()
```

```python
vendas_correta.query('Nome_Departamento == "Esporte_e_Lazer" and estado == "SP"').describe()
```

```python
# Os mais vendidos por Departamentos
vendas_correta.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correta.groupby('Nome_Departamento')['Preo_com_frete'].agg('mean').sort_values(ascending=False).reset_index()
```

```python
vendas_correta.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correta['Mes'] = vendas_correta['Data'].dt.month
vendas_correta.head()
```

```python
vendas_correta.groupby('Mes').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
cliente_vendas = vendas_correta.merge(cliente, how='left', on='cliente_Log')
cliente_vendas.head()
```

```python
agg_idcanalvenda_renda = round(cliente_vendas.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanalvenda_renda
```

```python
agg_idade_bandeira = round(cliente_vendas.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),0)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,8))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')

plt.ylabel('Mdia de Idade')

plt.title('Idade mdia por Bandeira', loc='center', fontsize=18, fontweight='bold', color='black')

plt.show();
```

```python
fig = plt.figure(figsize=(10,8))

plt.bar(agg_idcanalvenda_renda['idcanalvenda'], agg_idcanalvenda_renda['renda'], color='green')

plt.ylabel('Mdia de Renda')

plt.title('Renda mdia por Canal de venda', loc='center', fontsize=18, fontweight='bold', color='black')

plt.show();
```

```python
venda_por_data = vendas_correta.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(15,5))

plt.xlabel('Data da Venda')
plt.ylabel('Quantidade de venda')

x = venda_por_data['Data']
y = venda_por_data['idcompra']

plt.plot(x,y, color='red', linewidth = '2')

plt.show();
```

```python
vendas_departamento = round(vendas_correta.groupby('Nome_Departamento')['Preo_com_frete'].agg('mean'),2).sort_values(ascending=False).reset_index()
vendas_departamento
```

```python
fig = plt.figure(figsize=(40,10))

plt.bar(vendas_departamento['Nome_Departamento'], vendas_departamento['Preo_com_frete'], color='green')

plt.ylabel('Mdia de Preo')

plt.title('Preo mdio por Departamento', loc='center', fontsize=18, fontweight='bold', color='black')

plt.show();
```
